Desenvolva um programa que recebe do usuário nome completo e ano de nascimento que seja entre 1922 e 2021.
A partir dessas informações, o sistema mostrará o nome do usuário e a idade que completou, ou completará, no ano atual (2022).

Caso o usuário não digite um número ou apareça um inválido no campo do ano, o sistema informará o erro e continuará perguntando até que um valor correto seja preenchido.

```python
def ano_valido():
    while True:
        try:
            ano = int(input("Digite seu ano de nascimento entre 1922 e 2022: "))
            if 1922 <= ano <= 2022:
                return ano
            else: 
                print("Ano não se enquadra no parâmetro de idade aceito. Tente novamente.")
        except ValueError:
            print ("Valor inválido. Digite um valor que se enquadre nos parâmetros")

def main ():
    nome = input("Digite seu nome completo: ")
    ano = ano_valido()
    ano_atual = 2023 ### ano atual fornecido
    idade = ano_atual - ano

    print(f"Nome: {nome}")
    print(f"Idade: {idade} anos")

if __name__ == "__main__":
    main()


```
